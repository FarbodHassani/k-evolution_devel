## Fitting to Ngrid^2/3

```python
data = np.zeros((7,2))
for i in range(7):
    data[i,0] = Ngrid_solo[i]; 
    data[i,1] = 1./(kess_nl[i][-1:,2])-1

from scipy.optimize import curve_fit
def func(x, alpha, beta):
    return alpha *x + beta
x = data[:,0]**(2./3.)
y = 1+data[:,1]
init_vals = [1, 1]  
best_vals, covar = curve_fit(func, x, y, p0=init_vals)
```
