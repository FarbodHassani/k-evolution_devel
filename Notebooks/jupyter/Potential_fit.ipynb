```python
redshifts = [100.0,50.0,20.0,10.0 ,6.0,5.0,4.0,3.0,2.0,1.0,0.5,0.4,0.3,0.2,0.1,0.05,0.0];
class_files = [] 
for i in range(np.shape(redshifts)[0]):
    class_files.append(np.loadtxt('./CLASS_lcdm/lcdm_z'+str(i+1)+'_pk.dat'))
#     class_files = np.loadtxt('./CLASS_lcdm//lcdm_z1_pk.dat')
```

```python
x = np.arange(1.e-4,10)
# fig, ax = plt.subplots()

plt.figure(figsize=(14,10))

# plt.tick_params(axis='both', which='major', labelsize=10)
# plt.tick_params(axis='both', which='minor', labelsize=8)
for i in [1,3,4,5,7]:
    plt.plot(class_files[i][:,0], class_files[i][:,1],lw=3)

plt.plot(x,0.9 * x**(-5/2),lw=3,color="black",label = "$y = 0.9 k^{-5/2}$")
plt.plot(x,9. * x**(-5/2),lw=3,color="black",label = "$y = 9 k^{-5/2}$")
plt.plot(x,0.05 * x**(-5/2),lw=3,color="black",label = "$y = 0.05 k^{-5/2}$")


plt.yscale('log')
plt.xscale('log')

plt.xlim(1.e-4,5)
plt.ylim(1.e-3,1.e4)


plt.tick_params(axis='both', which='major', labelsize=20)
plt.tick_params(axis='both', which='minor', labelsize=20)


plt.legend(fontsize=15)
plt.ylabel('$P_m(k) \,[Mpc/h]^3$',fontsize=22)
plt.xlabel('$k[Mpc/h]$',fontsize=22)

```
