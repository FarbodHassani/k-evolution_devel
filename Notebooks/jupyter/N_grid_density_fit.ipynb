## Fitting

```python
data = np.zeros((5,2))
Ngrid = [32,64,128,258,512]
Ngrid=np.array(Ngrid)
for i in range(5):
    data[i,0] = Ngrid[i]; 
    data[i,1] = (maximum[name[i]]-avg[name[i]])/avg[name[i]]

from scipy.optimize import curve_fit
def func(x, alpha, beta):
    return alpha *x  +beta
x = np.log(data[:,0])
y = np.log(data[:,1])
init_vals = [1, 1]  
best_vals_density, covar = curve_fit(func, x, y, p0=init_vals)
```
