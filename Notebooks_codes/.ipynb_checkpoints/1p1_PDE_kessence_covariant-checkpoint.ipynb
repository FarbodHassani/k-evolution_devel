```python
import numpy as np
import matplotlib.pyplot as plt
```

# Scalar field

```python
class ScalarField(object):
    """
    Class for scalar fields.

    """

    def __init__(self,
                 n_grid=2**12,
                 ic_names=['sf', 'sf_p'],
                 ic_types={'sf':'flat', 'sf_p':'flat'},
                 ic_params={'sf':[1.], 'sf_p':[1.]}):
        # Flags
        self.is_ic_set = False
        self.is_run = False
        # Parameters for initial conditions
        self.n_grid = n_grid
        self.ic_names = ic_names
        self.ic_types = ic_types
        self.ic_params = ic_params
        # Placeholders
        self.ics = {key:None for key in self.ic_names}
        self.data = {}
        return
    
    def _check_types(self):
        assert isinstance(self.n_grid, int), 'n_grid should be an integer with the size of the grid on the spatial dimension'
        assert isinstance(self.ic_names, list), 'ic_names should be a list containing the names of the fields for which we need IC'
        assert isinstance(self.ic_types, dict), 'ic_types should be a dictionary with name_of_field:type_of_ic'
        assert isinstance(self.ic_params, dict), 'ic_params should be a dictionary with name_of_field:list_of_params'
        for nm in self.ic_names:
            assert nm in self.ic_types.keys(), 'Key {} not found in ic_types'.format(nm)
            assert nm in self.ic_params.keys(), 'Key {} not found in ic_params'.format(nm)
            assert isinstance(self.ic_types[nm], str), 'Value of ic_types[{}] should be string'.format(nm)
            assert isinstance(self.ic_params[nm], list), 'Value of ic_params[{}] should be list'.format(nm)
            assert all(isinstance(x, (int,float)) for x in self.ic_params[nm]), 'Values in the ic_params[{}] list should be float/integer'.format(nm)
    
    def _get_ic_flat(self,p):
        ic = p[0]*np.ones(self.n_grid)
        return ic

    def _get_ic_gaussian(self,p):
        ng = self.n_grid
        rg = np.arange(ng)
        ic = p[0]*np.float128(np.exp(-5.0*(2.0* np.float128(np.pi)*(rg - ng/2.)/ng)**2))
        return ic


    def set_ic(self,
               ic_types=None,
               ic_params=None,
               verbose=True):
        """
        Set the initial conditions for the scalar field and its derivative.
        """
        
        # Local variables
        ic_types_coded = ['flat', 'gaussian'] # Types of IC coded
        ic_n_params = [1, 1] # Number of parameters expected for each type if IC

        # If new parameters are specified overwrite default ones
        if ic_types:
            self.ic_types = ic_types
        if ic_params:
            self.ic_params = ic_params
        
        # Do the necessary checks
        self._check_types()
        for nm in self.ic_names:
            msg = 'Type of IC implemented: {}. For field {} provided {}'.format(ic_types_coded,nm,self.ic_types[nm])
            assert self.ic_types[nm] in ic_types_coded, msg
            idx = ic_types_coded.index(self.ic_types[nm])
            msg = 'Wrong number of parameters ({}) found for the field {} (expected {})'.format(len(self.ic_params[nm]),nm,ic_n_params[idx])
            assert len(self.ic_params[nm]) == ic_n_params[idx], msg
        
        # Get IC for each field
        for nm in self.ic_names:
            if self.ic_types[nm] == 'flat':
                self.ics[nm] = self._get_ic_flat(self.ic_params[nm])
            elif self.ic_types[nm] == 'gaussian':
                self.ics[nm] = self._get_ic_gaussian(self.ic_params[nm])
        
        # Print info and plots
        if verbose:
            for nm in self.ic_names:
                plt.plot(self.ics[nm],"*")
                plt.ylabel(nm,size=14)
                plt.show()
                is_sym = np.sum(self.ics[nm] - np.flip(self.ics[nm]))
                print('Checking symmetry of IC for field {}: {} (expected 0.)'.format(nm,is_sym))
        
        self.is_ic_set = True
        
        return
    
    def evolve_pde(self,
                   get_eqs,
                   params,
                   tau_ini=0.,
                   tau_end=250.,
                   dt=0.1,
                   dx_factor=2.0,
                   N_prints = 30,
                   force_eval=False,
                   verbose=True):
        
        assert self.is_ic_set, 'Initial conditions not yet provided. Use the set_ic method'
        assert (not self.is_run) or force_eval, 'Use the option force_eval if you want to evaluate again this pde'


        # Local variables
        a = 1 # TODO_EB: consider implementing FRW
        Hconf = 0
        tau_write = (tau_end - tau_ini)/N_prints
        dx=dx_factor*np.pi/self.n_grid
        num = 0

        # Data to store
        self.data = {
            'pi' : np.zeros((N_prints,self.n_grid)),
            'dpi_dt' : np.zeros((N_prints,self.n_grid)),
            'd2pi_dt2' : np.zeros((N_prints,self.n_grid)),
            'dpi_dx' : np.zeros((N_prints,self.n_grid)),
            'd2pi_dx2' : np.zeros((N_prints,self.n_grid)),
            'd2pi_dxdt' : np.zeros((N_prints,self.n_grid)),
            'tau' : np.zeros((N_prints,self.n_grid)),
            'detmet' : np.zeros((N_prints,self.n_grid)),
            'X' : np.zeros((N_prints,self.n_grid)),
        }

        # Initial conditions
        tau = tau_ini
        pi   = self.ics['sf']
        dpi_dt = self.ics['sf_p']

        # Simulation 1+1 D
        while (tau < tau_end and np.abs(np.max(pi))<1.e14):

            ######Rolling#########
            pi_plus  = np.roll(pi,-1)
            pi_2plus  = np.roll(pi,-2)
            pi_minus = np.roll(pi,1)
            pi_2minus = np.roll(pi,2)
            dpi_dt_plus = np.roll(dpi_dt,-1)
            dpi_dt_2plus = np.roll(dpi_dt,-2)
            dpi_dt_minus = np.roll(dpi_dt,1)
            dpi_dt_2minus = np.roll(dpi_dt,2)

            d2pi_dx2 = ((-pi_2plus + 16*pi_plus -30.0*pi + 16*pi_minus-pi_2minus)/(12.0*dx**2));
            dpi_dx = (pi_2minus - pi_2plus - 8*pi_minus + 8*pi_plus -0.*pi)/(12.0*dx);
            d2pi_dxdt = (dpi_dt_2minus - dpi_dt_2plus - 8*dpi_dt_minus + 8*dpi_dt_plus - 0.*dpi_dt)/(12.0*dx);

            #### Get theory ###
            X = (dpi_dt*dpi_dt - dpi_dx * dpi_dx )/(2.*a); # Kinetic term
            K, dK_dX, d2K_dX2, dK_dpi, d2K_dpidX = get_eqs(pi, X, params)

            #############  PDE solve #########
            term1 = a**2 * (dK_dpi  - dpi_dt**2 * d2K_dpidX/a**2 );
            term2 = - Hconf * (2.0 * dK_dX - dpi_dt**2 * d2K_dX2/a**2 ) * dpi_dt;
            term3 = dK_dX * d2pi_dx2
            term4 = d2K_dpidX * dpi_dx * dpi_dx;
            term5 = 2.0 * d2K_dX2 * dpi_dt * dpi_dx * d2pi_dxdt/a**2;
            term6 = -Hconf * d2K_dX2 * dpi_dt * dpi_dx * dpi_dx/a**2;
            term7 = - d2K_dX2 * dpi_dx * dpi_dx * d2pi_dx2/a**2;
    
            RHS = term1 + term2 + term3 + term4 +term5 +term6 + term7;
    
            d2pi_dt2 = RHS/(dK_dX + dpi_dt**2 * d2K_dX2/a**2);


            # Store results
            if (tau==tau_ini or np.abs(int(tau/tau_write)-tau/tau_write)<dt/tau_write or tau==tau_end):
                self.data['pi'][num,:] = pi
                self.data['dpi_dt'][num,:] = dpi_dt
                self.data['d2pi_dt2'][num,:] = d2pi_dt2
                self.data['dpi_dx'][num,:] = dpi_dx
                self.data['d2pi_dx2'][num,:] = d2pi_dx2
                self.data['d2pi_dxdt'][num,:] = d2pi_dxdt
                self.data['tau'][num,:] = tau
                self.data['X'][num,:] = X
                self.data['detmet'][num,:] = 1. + 2.*X*d2K_dX2/dK_dX
                
                if verbose:
                    print('{}/{} - Storing scalar field at tau = {:.2f}'.format(num+1,N_prints,tau))

                num+=1;

            # Euler algorithm
            dpi_dt = dpi_dt + dt*(d2pi_dt2); # pi'
            pi  = pi + dt * (dpi_dt) # pi
            tau = tau + dt
            if verbose and np.abs(np.max(pi))>=1.e6:
                print('Probable divergence starting at t_b = {:.2f}. pi_max = {:.0e}'.format(tau,pi.max()))

        return
    
    def plot_sf(self, fields, time_slices, figshape, figsize=(20,10)):
        
        assert len(figshape) == 2, 'figshape is not a valid shape!'
        assert figshape[0]*figshape[1] >= len(fields), 'You want to plot {} fields in a {} grid. That is impossible'.format(len(fields),figshape)

        plt.figure(figsize=figsize)
        
        for nf, f in enumerate(fields):
            plt.subplot(figshape[0],figshape[1],nf+1)
            for ts in time_slices:
                plt.plot(self.data[f][ts,:],"-",markersize=16,label=r"$\tau$ = {:.2e}".format(sf.data['tau'][ts][0]))
                plt.tick_params(labelsize=20)
                plt.title(f,fontsize=25)
                if nf==0:
                    plt.legend(fontsize=16)
        return
```

# Solutions

```python
def get_theory(pi, X, params):
    X_hat, g0, g2, g4 = params
    K = -g0 + g2 * (X - X_hat)**2 + g4 * (X - X_hat)**4 # K(X,phi)
    dK_dX = 2.0 * g2 * (X - X_hat) + 4.0 * g4 * (X - X_hat)**3
    d2K_dX2 = 2.0 * g2 + 12. * g4 * (X - X_hat)**2
    dK_dpi = 0
    d2K_dpidX =0
    return K, dK_dX, d2K_dX2, dK_dpi, d2K_dpidX


# Theory parameters
#            X_hat,      g0,      g2,      g4
params = [     8.e0,     0.0,     1.0, 1.e-1]

sf = ScalarField()
sf.set_ic(ic_types={'sf':'gaussian','sf_p':'flat'},ic_params={'sf':[-1.],'sf_p':[0.12]},verbose=False)
sf.evolve_pde(get_theory, params, tau_end=0.13, dt=0.000001,N_prints = 3000,verbose=True)
```
